```python
model_res = ResNet50(weights="imagenet", input_shape=(IMAGE_SIZE,IMAGE_SIZE,3))
```

```python
model_res.summary()
```

```python
from keras.models import Model, load_model
from keras.layers import Dense
```

```python
model_new = Model(model_res.input, model_res.layers[-2].output)
```

```python
x = Dense(5, activation='sigmoid')(model_new.output)
model_new = Model(model_new.input, x)
model_new.summary()
```

```python
model_new.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False),
    metrics = ['accuracy']
)
```

```python
history_2 = model_new.fit(
    dataset_train,
    batch_size = BATCH_SIZE,
    verbose=1,
    epochs = 15
)
```

```python
scores = model_new.evaluate(dataset_test)
```
